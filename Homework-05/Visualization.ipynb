```python
import pickle
import os
import matplotlib.pyplot as plt
FIGURESIZE=(20,15)
FONTSIZE=18
plt.rcParams['figure.figsize'] = FIGURESIZE
plt.rcParams['font.size'] = FONTSIZE
plt.rcParams['xtick.labelsize'] = FONTSIZE
plt.rcParams['ytick.labelsize'] = FONTSIZE
```

```python
data1 = pickle.load(open(os.path.join('results_fold_1.pkl'),'rb'))
data2 = pickle.load(open(os.path.join('results_fold_2.pkl'),'rb'))
data3 = pickle.load(open(os.path.join('results_fold_3.pkl'),'rb'))
data4 = pickle.load(open(os.path.join('results_fold_4.pkl'),'rb'))
data5 = pickle.load(open(os.path.join('results_fold_5.pkl'),'rb'))
```

```python
data3['predict_testing_eval']
```

## Figure I - Training AUC vs Epochs

```python
plt.plot(data1['history']['AUC'], linestyle='solid',label='fold_%d'%1)
plt.plot(data2['history']['AUC'], linestyle='solid',label='fold_%d'%2)
plt.plot(data3['history']['AUC'], linestyle='solid',label='fold_%d'%3)
plt.plot(data4['history']['AUC'], linestyle='solid',label='fold_%d'%4)
plt.plot(data5['history']['AUC'], linestyle='solid',label='fold_%d'%5)

plt.ylabel('Training AUC')
plt.xlabel('Epochs')
plt.title('Training AUCs vs Epochs')
plt.legend()
plt.savefig('Figure_I.jpg',bbox_inches='tight', dpi=150)
```

## Figure II - Validation AUC vs Epochs

```python
plt.plot(data1['history']['val_AUC'], linestyle='solid',label='fold_%d'%1)
plt.plot(data2['history']['val_AUC'], linestyle='solid',label='fold_%d'%2)
plt.plot(data3['history']['val_AUC'], linestyle='solid',label='fold_%d'%3)
plt.plot(data4['history']['val_AUC'], linestyle='solid',label='fold_%d'%4)
plt.plot(data5['history']['val_AUC'], linestyle='solid',label='fold_%d'%5)

plt.ylabel('Validation AUC')
plt.xlabel('Epochs')
plt.title('Validation AUCs vs Epochs')
plt.legend()
plt.savefig('Figure_II.jpg',bbox_inches='tight', dpi=150)
```

## Figure III - Training Set Accuracy vs Epochs

```python
plt.plot(data1['history']['binary_accuracy'], linestyle='solid',label='fold_%d'%1)
plt.plot(data2['history']['binary_accuracy'], linestyle='solid',label='fold_%d'%2)
plt.plot(data3['history']['binary_accuracy'], linestyle='solid',label='fold_%d'%3)
plt.plot(data4['history']['binary_accuracy'], linestyle='solid',label='fold_%d'%4)
plt.plot(data5['history']['binary_accuracy'], linestyle='solid',label='fold_%d'%5)

plt.ylabel('Training Set Accuracy')
plt.xlabel('Epochs')
plt.title('Training Set Accuracy vs Epochs')
plt.legend()
plt.savefig('Figure_III.jpg',bbox_inches='tight', dpi=150)
```

## Figure IV - Validation Set Accuracy vs Epochs

```python
plt.plot(data1['history']['val_binary_accuracy'], linestyle='solid',label='fold_%d'%1)
plt.plot(data2['history']['val_binary_accuracy'], linestyle='solid',label='fold_%d'%2)
plt.plot(data3['history']['val_binary_accuracy'], linestyle='solid',label='fold_%d'%3)
plt.plot(data4['history']['val_binary_accuracy'], linestyle='solid',label='fold_%d'%4)
plt.plot(data5['history']['val_binary_accuracy'], linestyle='solid',label='fold_%d'%5)

plt.ylabel('Validation Set Accuracy')
plt.xlabel('Epochs')
plt.title('Validation Set Accuracy vs Epochs')
plt.legend()
plt.savefig('Figure_IV.jpg',bbox_inches='tight', dpi=150)
```

## Figure V - Histogram of Test AUC

```python
import numpy as np
l = [data1['predict_testing_eval'][-1],data2['predict_testing_eval'][-1],data3['predict_testing_eval'][-1],data4['predict_testing_eval'][-1],data5['predict_testing_eval'][-1]]
plt.hist(l,alpha=0.5)
plt.axvline(np.mean(l), color='k',linestyle='dashed', linewidth=1)

min_ylim, max_ylim = plt.ylim()
plt.text(np.mean(l), max_ylim, 'Mean Test AUC : {:.2f}'.format(np.mean(l)))
plt.xlabel('Test Set AUC')
plt.savefig('Figure_V.jpg',bbox_inches='tight', dpi=150)
```
